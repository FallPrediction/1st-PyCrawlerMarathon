## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
r.json()
```

```python
data=r.json()
print('這個 API 一次會回傳%d筆資料' % len(data))
print('每一筆資料包含%s欄位' % data[0].keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
for d in data:
    print('標題 %s' % d['title'])
    print('貼文時間 %s' % d['createdAt'])
    print('留言人數 %s' % d['commentCount'])
    print('按讚人數 %s' % d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
import pandas as pd
```

```python
popular_df=pd.read_json(r.text, orient='records')
print('熱門文章的平均留言人數', popular_df['commentCount'].mean())
print('熱門文章的平均按讚人數', popular_df['likeCount'].mean())
```

```python
r=requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
unpopular_df=pd.read_json(r.text, orient='records')
print('非熱門文章的平均留言人數', unpopular_df['commentCount'].mean())
print('非熱門文章的平均按讚人數', unpopular_df['likeCount'].mean())
```
