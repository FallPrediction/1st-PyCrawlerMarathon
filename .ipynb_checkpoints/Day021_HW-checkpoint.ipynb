## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
import datetime
import time
browser = webdriver.Chrome(executable_path='chromedriver')
```

```python
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

#今天午夜
midnight=datetime.datetime.combine(datetime.date.today(), datetime.time.min)
for d in soup.find(class_="part_list_2").find_all('h3'):
    part_date=d.find(class_="date").text
    if datetime.datetime.strptime(part_date, "%Y/%m/%d %H:%M").date()==datetime.datetime.today().date():
        print(part_date, d.find_all('a')[-1].text)
    else:
        break
```

## 2. 取出現在時間兩小時內的新聞

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    part_date=datetime.datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if part_date >= datetime.datetime.now()- datetime.timedelta(hours=2):
        print(part_date, d.find_all('a')[-1].text)
    else:
        break
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
three_days_ago=datetime.date.today()- datetime.timedelta(days=3)
year=datetime.datetime.strftime(three_days_ago, "%Y")
month=datetime.datetime.strftime(three_days_ago, "%m")
day=datetime.datetime.strftime(three_days_ago, "%d")
print(year, month, day)
```

```python
#查詢前三天的新聞
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
from selenium.webdriver.common.by import By

selY=WebDriverWait(browser, 10).until(ec.element_to_be_clickable((By.ID, 'selY')))
selY.send_keys(year)
selM=WebDriverWait(browser, 10).until(ec.element_to_be_clickable((By.ID, 'selM')))
selM.send_keys(month)
selD=WebDriverWait(browser, 10).until(ec.element_to_be_clickable((By.ID, 'selD')))
selD.send_keys(day)
button=WebDriverWait(browser, 10).until(ec.element_to_be_clickable((By.ID, 'button')))
button.click()
```

```python
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for d in soup.find(class_="part_list_2").find_all('h3'):
    part_date=datetime.datetime.strptime(d.find(class_="date").text, "%Y/%m/%d %H:%M")
    if part_date >= datetime.datetime.combine(three_days_ago, datetime.time(15)) and part_date <= datetime.datetime.combine(three_days_ago, datetime.time(17)):
        print(part_date, d.find_all('a')[-1].text)
```
